```python
import matplotlib.pyplot as plt
import random
import math as math

# estimate value of pi using Monte Carlo simulation
def estimate_pi(n):
    inside_pts = 0
    for _ in range(n):
        x = random.uniform(0,1)
        y = random.uniform(0,1)
        if math.sqrt(x**2 + y**2) <= 1:
            inside_pts += 1
    est_pi = 4 * inside_pts/n
    error = abs((math.pi - est_pi)/math.pi) * 100
    return (est_pi, error)

# returns points to plot the estimates of pi
def plot_estimate(n):
    in_pts = []
    out_pts = []
    inside_pts = 0
    for _ in range(n):
        point = (random.uniform(0,1), random.uniform(0,1))
        if math.sqrt(point[0]**2 + point[1]**2) <= 1:
            inside_pts += 1
            in_pts.append(point)
        else:
            out_pts.append(point)
    est_pi = 4 * inside_pts/n
    error = abs((math.pi - est_pi)/math.pi) * 100
    return [est_pi, error, in_pts, out_pts]
    

if __name__ == "__main__":
    n_list = [] # creates list of n^3,4,5,6
    estimate_n4 = []
    for i in range(3,7):
        n_list.append(pow(10,i))
    
    for j in n_list:
        #to plot 10^4
        if n_list.index(j) == 1:
            estimate_n4 = plot_estimate(j)
            print("n = 10 ^ {0} pi = {1:.6f} error = {2:.4f} %".format(n_list.index(j)+3,estimate_n4[0],estimate_n4[1])) 
        else:
            estimate = estimate_pi(j)
            print("n = 10 ^ {0} pi = {1:.6f} error = {2:.4f} %".format(n_list.index(j)+3,estimate[0],estimate[1]))
    
    print("\nPlotting for 10^4:")
    plt.figure(figsize=(6,6))
    in_x = [i[0] for i in estimate_n4[2]]
    in_y = [i[1] for i in estimate_n4[2]]
    out_x = [i[0] for i in estimate_n4[3]]
    out_y = [i[1] for i in estimate_n4[3]]
    plt.scatter(in_x, in_y, 1, marker='.', color='b')
    plt.scatter(out_x, out_y, 1, marker='.', color='r')
    plt.title('Figure 1: Estimating  using simulation')
    plt.show()
      
```
