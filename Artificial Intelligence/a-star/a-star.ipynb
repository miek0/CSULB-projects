```python
# CECS 451 
# Assignment 2 - A* Algorithm
# Mike Zeng

map_txt = open("map.txt", "r")
maps = map_txt.read().split('\n')
map_dist = []
for i in maps:
    start = i.split('-')
    dest = start[1].split(',')
    length = len(dest)
    tup = tuple(start[0] if j < 1 else (dest[j-1][0:dest[j-1].find('(')], int(dest[j-1][dest[j-1].find('(')+1:dest[j-1].rfind(')')])) for j in range(len(dest)+1))
    map_dist.append(tup)
    
dist_txt = open("distances.txt","r")
dists = dist_txt.read().split('\n')
city_dist = []
for i in dists:
    city = i.split('-')
    city_dist.append((city[0],int(city[1])))
    
dist_txt.close()
map_txt.close()

class City():
    def __init__(self, parent, next_city):
        self.parent = parent
        self.name = next_city[0]
        self.position = next_city[1]

        self.f = 0
        self.g = 0
        self.h = 0

    def __eq__(self, other):
        return self.position == other.position

def astar(start, end):
    start_city = City(None, start)
    start_city.g = start_city.h = start_city.f = 0
    end_city = City(None, end)
    end_city.g = end_city.h = end_city.f = 0

    not_reached_list = []
    reached_list = []
    
    not_reached_list.append(start_city)
    
    while len(not_reached_list) > 0:
        current_city = not_reached_list[0]
        current_index = 0
        for index, item in enumerate(not_reached_list):
            if item.f < current_city.f:
                current_city = item
                current_index = index

        not_reached_list.pop(current_index)
        reached_list.append(current_city)

        if current_city.name == end_city.name:
            path = []
            current = current_city
            while current is not None:
                path.append(current)
                current = current.parent
            return path[::-1]
        
        children = []
        adjacent = []
        for m in map_dist:
            if current_city.name == m[0]:
                for n in m:
                    adjacent.append(n)
        adjacent = adjacent[1:]
        
        for new_position in adjacent:
            new_city = City(current_city, (new_position[0], new_position[1]))
            new_city.g = new_position[1]
            children.append(new_city)

        for child in children:
            for closed_child in reached_list:
                if child == closed_child:
                    continue
            child.g += current_city.g
            for i in city_dist:
                if child.name == i[0]:
                    child.h = i[1]
            child.f = child.g + child.h
            for open_city in not_reached_list:
                if child == open_city and child.g > open_city.g:
                    continue
            not_reached_list.append(child)
        
def main():
    start_location = input("Where do you want to begin? ")
    start_location = start_location.lower().capitalize()
    if start_location==city_dist[14][0].lower().capitalize():
        start_location = start_location.replace('v','V')
    end = city_dist[1]
    start = ()
    for begin in city_dist:
        if begin[0] == start_location:
            start = (start_location, begin[1])
    path = astar(start, end)
    print("From city: ", start_location)
    print("To city: ", end[0])
    path_names = []
    for i in path:
        path_names.append(i.name)
    print("Best route: ", path_names)
    print("Total distance: ", path[-1].g)

if __name__ == '__main__':
    main()
```
