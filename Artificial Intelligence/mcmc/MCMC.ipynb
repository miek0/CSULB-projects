```python
import random

## Main of the program.
if __name__ == "__main__":
  ## Indicate we're doing Part A first.  I'm hardcoding them into tuples because I thought we'd use the values later.
  print("Part A. The sampling probabilities")
  ## P(C|-s, r) is hardcoded, I did it separately.
  Cnsr = (0.8781, 0.1220)
  ## Print the results, formatting with 4 decimal spaces.
  print("P(C|-s, r) = <{0:.4f}, {1:.4f}>".format(Cnsr[0], Cnsr[1]))
  ## P(C|-s, -r) is hardcoded.
  Cnsnr = (0.3103, 0.6900)
  ## Print the results the way you want them again.
  print("P(C|-s, -r) = <{0:.4f}, {1:.4f}>".format(Cnsnr[0], Cnsnr[1]))
  ## I figure you get the gist of what I'm doing by now.
  cnsRw = (0.9887, 0.0130)
  print("P(R|c, -s, w) = <{0:.4f}, {1:.4f}>".format(cnsRw[0], cnsRw[1]))
  ## Same here, the principle is pretty redundant.
  ncnsRw = (0.8182, 0.1818)
  print("P(R|-c, -s, w) = <{0:.4f}, {1:.4f}>".format(ncnsRw[0], ncnsRw[1]) + "\n")
  
  ## Oh hey, part B.  I also hardcoded these.  I shoved 'em into lists to use 'em later.
  print("Part B. The transition probability matrix")
  ## Just following the template.
  print("\tS1\tS2\tS3\tS4")
  ## S1 to S1 is just 1 - S1 to S2 and S1 to S3.  I'd figure out the fancy math but it's really not needed for this particular problem.
  ## S1 to S2 is 0.5 * P(-r|c, -s, w)
  ## S1 to S3 is 0.5 * P(-c|-s, r)
  ## S1 to S4 is impossible.
  S1Trans = [0.9321, 0.0069, 0.0610, 0.0000]
  print("S1\t{0:.4f}\t{1:.4f}\t{2:.4f}\t{3:.4f}".format(S1Trans[0], S1Trans[1], S1Trans[2], S1Trans[3]))
  ## S2 to S1 is 0.5 * P(r|c, -s, w)
  ## S2 to S2 is 1 - S2 to S1 and S2 to S4
  ## S2 to S3 is impossible.
  ## S2 to S4 is 0.5 * P(-c|-s, -r)
  S2Trans = [0.4932, 0.1619, 0.0000, 0.3449]
  print("S1\t{0:.4f}\t{1:.4f}\t{2:.4f}\t{3:.4f}".format(S2Trans[0], S2Trans[1], S2Trans[2], S2Trans[3]))
  ## S3 to S1 is 0.5 * P(c|-s, r)
  ## S3 to S2 is impossible.
  ## S3 to S3 is 1 - S3 to S1 and S3 to S4
  ## S3 to S4 is 0.5 * P(-r|-c, -s, w)
  S3Trans = [0.4391, 0.0000, 0.4700, 0.0909]
  print("S1\t{0:.4f}\t{1:.4f}\t{2:.4f}\t{3:.4f}".format(S3Trans[0], S3Trans[1], S3Trans[2], S3Trans[3]))
  ## S4 to S1 is impossible.
  ## S4 to S2 is 0.5 * P(c|-s, -r)
  ## S4 to S3 is 0.5 * P(r|-c, -s, w)
  ## S4 to S4 is 1 - S4 to S2 and S4 to S3
  S4Trans = [0.0000, 0.1552, 0.4091, 0.4357]
  print("S1\t{0:.4f}\t{1:.4f}\t{2:.4f}\t{3:.4f}".format(S4Trans[0], S4Trans[1], S4Trans[2], S4Trans[3]) + "\n")

  ## I kinda hardcoded this?  Not really?  I mean it's calculated in here but I set it only to work for this prompt.
  print("Part C. The Probability for the query")
  ## Initialize a CSRW state counter, with C and R randomized.
  current = [random.randint(0, 1), 0, random.randint(0, 1), 1]
  ## Initialize two counters that track C.
  ## Counter for when C is true.
  c = 0
  ## Counter for when C is false.
  nc = 0
  ## Run through this 1,000,000 times:
  for iteration in range(1000000):
    ## Generate a random transition number from 0 to 1.
    transition = random.uniform(0, 1)
    ## Get ready for some monkey brain BS.  I hardcoded what's gonna happen for every possible state, so 12 different possibilities.  God, the headache of fixing it was annoying.
    ## We're currently in State 1:
    if current == [1, 0, 1, 1]:
      ## If the transition generated is within the S1 to S1 bounds:
      if transition < S1Trans[0]:
        ## Append onto the C true counter.
        c += 1
        ## Stay here.
      ## If the transition generated is within the S1 to S2 bounds:
      elif transition < S1Trans[0] + S1Trans[1]:
        ## Append onto the C true counter.
        c += 1
        ## Move onto to State 2
        current = [1, 0, 0, 1]
      ## If the transition generated is within the S1 to S3 bounds:
      elif transition < S1Trans[0] + S1Trans[1] + S1Trans[2]:
        ## Append onto the C false counter.
        nc += 1
        ## Move onto State 3.
        current = [0, 0, 1, 1]
      ## It's impossible to move onto State 4 so I'm not gonna code it.
    ## We're currently in State 2:
    elif current == [1, 0, 0, 1]:
      ## If the transition generated is within the S2 to S1 bounds:
      if transition < S2Trans[0]:
        ## Append onto the C true counter.
        c += 1
        ## Move onto State 1
        current = [1, 0, 1, 1]
      ## If the transition generated is within the S2 to S2 bounds:
      elif transition < S2Trans[0] + S2Trans[1]:
        ## Append onto the C true counter.
        c += 1
        ## The current doesn't change.
      ## S2 can't reach S3 so I won't bother coding something for it.
      ## If the transition generated is within the S2 to S4 bounds:
      elif transition < S2Trans[0] + S2Trans[1] + S2Trans[3]:
        ## Append onto the C false counter.
        nc += 1
        ## Move onto State 4
        current = [0, 0, 0, 1]
    ## We're currently in State 3:
    elif current == [0, 0, 1, 1]:
      ## If the transition generated is within the S3 to S1 bounds:
      if transition < S3Trans[0]:
        ## Append onto the C true counter.
        c += 1
        ## Move onto State 1
        current = [1, 0, 1, 1]
      ## S3 can't reach S2 so I won't bother coding something for it.
      ## If the transition generated is within the S3 to S3 bounds:
      elif transition < S3Trans[0] + S3Trans[2]:
        ## Append onto the C false counter.
        nc += 1
        ## The current doesn't change.
      ## If the transition generated is within the S3 to S4 bounds:
      elif transition < S3Trans[0] + S3Trans[2] + S3Trans[3]:
        ## Append onto the C false counter.
        nc += 1
        ## Move onto State 4
        current = [0, 0, 0, 1]
    ## We're currently in State 4:
    elif current == [0, 0, 0, 1]:
      ## S4 can't reach S1 so I won't bother coding something for it.
      ## If the transition generated is within the S4 to S2 bounds:
      if transition < S4Trans[1]:
        ## Append onto the C true counter.
        c += 1
        ## Move onto State 2
        current = [1, 0, 0, 1]
      ## If the transition generated is within the S4 to S3 bounds:
      elif transition <S4Trans[1] + S4Trans[2]:
        ## Append onto the C false counter.
        nc += 1
        ## Move onto State 3.  Originally I had this SOB set to [0, 0, 0, 1], which was a typo and borked up my accuracy.  Took me half an hour to find it!
        current = [0, 0, 1, 1]
      ## If the transition generated is within the S4 to S4 bounds:
      elif transition <S4Trans[1] + S4Trans[2] + S4Trans[3]:
        ## Append onto the C false counter.
        nc += 1
        ## Stay in S4
  ## Divide by 1,000,000 to get the probability of C being true.
  c = c / 1000000
  ## Divide by 1,000,000 to get the probability of C being false.
  nc = nc / 1000000
  ## Print the formatted results.  Should be along the lines of <0.8566, 0.1434>.
  print("P(C|-s, w) = <{0:.4f}, {1:.4f}>".format(c, nc))
```
